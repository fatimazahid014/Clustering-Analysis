### Using Different Initialization Parameters to Achieve a Suitable  Outcome

```python
#@title Importing modules
import pandas as pd
from sklearn.cluster import KMeans
import altair as alt
import numpy as np
```

```python
#@title link to dataset
file_url = 'https://raw.githubusercontent.com/fenago/DSBook/main/Chapter%205/Mall_Customers.csv'
```

```python
#@title LOading dataset
data = pd.read_csv(file_url, \
                 usecols=['Age', \
                          'Annual Income (k$)', \
'Spending Score (1-100)'])
```

```python
data.head()
```

```python
#@title Assigning the columns to a new variable
X = data[['Age', \
        'Annual Income (k$)']]
```

```python
#@title Fit a k-means model with n_init equal to 1 and a random init
kmeans = KMeans(random_state=1, n_clusters=4, \
                init='random', n_init=1)
kmeans.fit(X)
```

```python
#@title Using the predict method from the sklearn package, predict the clustering assignment from the input variable, (X)
data['cluster3'] = kmeans.predict(X)
```

```python
#@title Plot the clusters using an interactive scatter plot. 
scatter_plot = alt.Chart(data).mark_circle()
```

```python
#@title Use the encode and interactive methods from altair to specify the display of the scatter plot and its interactivity options
scatter_plot.encode(x='Age', \
                    y='Annual Income (k$)', \
                    color='cluster3:N', \
                    tooltip=['Spending Score (1-100)', 'cluster3', \
                             'Age', \
'Annual Income (k$)'])\
                   .interactive()
```

```python
#@title Repeat Steps 5 to 8 but with different k-means hyperparameters, n_init=10 and random init,
kmeans = KMeans(random_state=1, n_clusters=4, \
                init='random', n_init=10)
kmeans.fit(X)
data['cluster4'] = kmeans.predict(X)
scatter_plot = alt.Chart(data).mark_circle()
scatter_plot.encode(x='Age', \
                    y='Annual Income (k$)', \
                    color='cluster4:N',
                    tooltip=['Spending Score (1-100)', 'cluster4', \
                             'Age', \
'Annual Income (k$)'])\
                   .interactive()
```

```python
#@title Again, repeat Steps 5 to 8 but with different k-means hyperparameters â€“ n_init=100 and random init
kmeans = KMeans(random_state=1, n_clusters=4, \
                init='random', n_init=100)
kmeans.fit(X)
data['cluster5'] = kmeans.predict(X)
scatter_plot = alt.Chart(data).mark_circle()
scatter_plot.encode(x='Age', \
                    y='Annual Income (k$)', \
                    color='cluster5:N', \
                    tooltip=['Spending Score (1-100)', 'cluster5', \
                    'Age', \
                    'Annual Income (k$)'])\
            .interactive()
```
