### Standardizing the Data from Our Dataset

```python
#@title importing modules
import pandas as pd
from sklearn.cluster import KMeans
import altair as alt
import numpy as np
```

```python
file_url = 'https://raw.githubusercontent.com/fenago/DSBook/main/Chapter%205/Mall_Customers.csv'
```

```python
#@title loading data
data = pd.read_csv(file_url, \
                 usecols=['Age', \
                          'Annual Income (k$)', \
'Spending Score (1-100)'])
```

```python
#@title displaying first 5 rows
data.head()
```

```python
#@title assigning extracted columns to a new variable.
X = data[['Age', \
        'Annual Income (k$)']]
```

```python
#@title Import the MinMaxScaler and StandardScaler classes from sklearn
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
#@title Instantiate and fit MinMaxScaler with the data:
min_max_scaler = MinMaxScaler()
min_max_scaler.fit(X)
```

```python
#@title Perform the min-max scaling transformation and save the data into a new variable called X_min_max
X_min_max = min_max_scaler.transform(X)
X_min_max
```

```python
#@title Fit a k-means model on the scaled data with the following hyperparameters: random_state=1, n_clusters=4, init='k-means++', n_init=5
kmeans = KMeans(random_state=1, n_clusters=4, \
                init='k-means++', n_init=5)
kmeans.fit(X_min_max)
```

```python
#@title Assign the k-means predictions of each value of X in a new column called 'cluster8' 
data['cluster8'] = kmeans.predict(X_min_max)
```

```python
#@title Plot the k-means results into a scatter plot using the altair package
scatter_plot = alt.Chart(data).mark_circle()
scatter_plot.encode(x='Age', \
                    y='Annual Income (k$)',\
                    color='cluster8:N',\
                    tooltip=['Spending Score (1-100)', 'cluster8', \
                             'Age',\
'Annual Income (k$)'])\
                   .interactive()
```

```python
#@title Re-train the k-means model but on the z-score-standardized data with the same hyperparameter values, random_state=1, n_clusters=4, init='kmeans++', n_init=5
standard_scaler = StandardScaler()
X_scaled = standard_scaler.fit_transform(X)
kmeans = KMeans(random_state=1, n_clusters=4, \
                init='k-means++', n_init=5)
kmeans.fit(X_scaled)

data['cluster9'] = kmeans.predict(X_scaled)
```

```python
#@title Plot the k-means results in a scatter plot using the altair package
scatter_plot = alt.Chart(data).mark_circle()
scatter_plot.encode(x='Age', \
                    y='Annual Income (k$)', \
                    color='cluster9:N', \
                    tooltip=['Spending Score (1-100)', 'cluster9', \
                             'Age',\
'Annual Income (k$)'])\
                   .interactive()
```
